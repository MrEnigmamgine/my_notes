# Tank Circuits

```python
L = 100*si.m
C = 5*si.p
R = 1*si.k
V = 10
f = get_resonant_frequency(L,C)

Xc = get_capacitive_reactance(C,f)
Xl = get_inductive_reactance(L,f)

Z_series = sum([R,Xl,Xc])
Z_parallel = R + recipsumrecip([Xl,Xc])

Z_series, Z_parallel
```

```python
abs(Z_series), abs(Z_parallel)
```

```python
vectors
```

```python
vectors = [n2v(n) for n in [Z_series,Z_parallel]]
# vectors = list(vectors)
fig, ax = viz.plot_vector_diagram([vectors[0]])
ax.xaxis.set_major_formatter(engFormat)
ax.yaxis.set_major_formatter(engFormat)
fig, ax = viz.plot_vector_diagram([(0,0),vectors[1]])
ax.xaxis.set_major_formatter(engFormat)
ax.yaxis.set_major_formatter(engFormat)
```
